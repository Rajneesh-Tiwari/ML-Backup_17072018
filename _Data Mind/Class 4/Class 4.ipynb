***Loss Functions***

- A function that evaluates the goodness of our models
- Mainly evaluates the difference between: $Y - \hat{Y}$    (Act - Pred) by assigning some kind of function
- Many are already implemented in python's ML library sklearn (EG: RMSE, binary/cross entropy loss etc)

### For Regression

- ${Mean Squared Error} = \frac{1}{N} \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2$

- ${Root Mean Squared Error} = \sqrt{\frac{1}{N} \sum_{i=1}^{n} (Y_i - \hat{Y_i})^2}$

- ${Mean Absolute Error} = \frac{1}{N} \sum_{i=1}^{n} |(Y_i - \hat{Y_i})|$

- ${Mean Absolute Percent Error} = \frac{100}{N}*\sum_{i=1}^{N} |\frac{Y_i- \hat{Y_i}}{Y_i}| $



- $Y_i$ = Target value for $i_{th}$ example
- $\hat{Y_i}$ = Predicted value for $i_{th}$ example


### For Classification

$ Cross Entropy Loss (Log Loss) = -\frac{1}{n} \sum_{i=1}^n Y_i \log(\hat Y_i) + (1-Y_i) log(1- \hat Y_i) $


- $Y_i$ = Target value for $i_{th}$ example
- $\hat{Y_i}$ = Predicted value for $i_{th}$ example
- We can also use some of the lossed defined above as well for classification, however, it is unusual to do so.

Some examples:

```python
from sklearn.metrics import mean_squared_error   #### import predefined loss metric
import numpy as np
np.random.seed(8)   #### set seed to replicate results

### make random data

y_true = np.array([4.5,5,8,23,12])
y_pred = np.array([i+np.random.random(1) for i in y_true])   ### adding a uniform random number to true targets

print("Mean squared error {}".format(mean_squared_error(y_true, y_pred)))
```

```python
#### fn for MAPE since it is not implemented as such in Sklearn
def MAPE(y_true, y_pred): ### both Ys are np arrays
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print("MAPE: {}%".format(np.round(MAPE(y_true, y_pred)),2))
```

```python
from sklearn.metrics import log_loss   #### import predefined loss metric
import numpy as np

### make random data
y_true = np.array([1,0,1,1,0])
y_pred = np.array([0.95,0.25,0.35,0.57,0.87])   ### These are probabilities of positive class or being a 1

print("Log loss {}".format(log_loss(y_true, y_pred)))
print("Mean squared error {}".format(mean_squared_error(y_true, y_pred)))   ### use regression loss for classification
```

***Linear algebra refresh***

- ** Scalars **
    - Magnitude
- **Vectors **
    - Magnitude and Angle
    - Addition, Subtraction, Multiplication
    - Dot product or element wise multiplication and sum
- ** Matrix**
    - N*T array

    - Transpose

```python
#### Vectors are 1D np arrays; 2D arrays are matrix or dataframes

np.random.seed(5) ### to make random numbers same
vec1 = np.random.random_integers(5,size=(5,1))
vec2 = np.random.random_integers(5,size=(5,1))
scalar = 2

print(vec1,'\n',"*"*50,'\n',vec2)
print('*'*50)
print("Vector sum is element wise sum \n",vec1+vec2)
print('*'*50)
print("Vector difference is element wise diff \n",vec1-vec2)
print('*'*50)
print("Vector multiplication is element wise multiplication \n",vec1*vec2)
print('*'*50)
print("Vector division is element wise division \n",vec1/vec2)
print('*'*50)
print("To get dot product b/w 2 1D arrays \n",np.dot(vec1.T,vec2))   ### remember rows of 1st vec times the cols to 2nd vec; hence the transpose
### print(vec1.T.dot(vec2))  ### alternative way to do dot pdt b/w 1D vectors
print('*'*50)
print("We can also get dot pdt by summing over the multiplication \n",np.sum(vec1*vec2))

### np.sum(vec1*vec2) == np.dot(vec1.T,vec2)  --- check this to make sure
```

```python
### Recall broadcasting which enables us to do math operations on different sized vectors by broadcasting 1 array to the size of other
print('*'*50)
print("Vector: \n",vec1)
print("*"*50)
print("Scalar: \n",scalar)
print('*'*50)
print("Scalar addition to vector takes advantage of broadcasting \n",scalar+vec1)
print('*'*50)
print("Scalar substraction to vector takes advantage of broadcasting \n",vec1-scalar)
print('*'*50)
print("Scalar multiplication to vector takes advantage of broadcasting \n",scalar*vec1)
print('*'*50)
print("Scalar division to vector takes advantage of broadcasting \n",vec1/scalar)
```

```python
### Matrix multiplication in numpy

mat1 = np.matrix(([4, 5, 6],[1, 6, 5],[3 ,4, 8],[2, 12, 3]))
mat2 = np.matrix(([3, 4, 6],[5, 6, 7],[6,56, 7]))

print('*'*50)
print("Mat1 shape: \n",mat1.shape)
print('*'*50)
print("Mat2 shape: \n",mat2.shape)
print("Columns of Mat1: {} = Rows of Mat 2: {}; hence mat multiplication is possible".format(mat1.shape[1],mat2.shape[0]))
print('*'*50)
print("Matrix multiplication via @ operator in Numpy: \n",mat1@mat2)   ### rem,rows of 1st times cols of 2nd
print("Matrix multiplication via dot pdt operator in numpy: \n",np.dot(mat1,mat2))   ### rows of 1st times cols of 2nd
print("Matrix multiplication via x.dot(y): \n",mat1.dot(mat2))   ### rows of 1st times cols of 2nd
```

```python
### matrix - vector operations
vec3 = np.array([1,4,2,5]).reshape(-1,1)  ### equal to number of rows in our matrix
print('*'*50)
print(mat1)
print('*'*50)
print(vec3)
print('*'*50)
print("Vector addition to matrix takes advantage of broadcasting \n",vec3+mat1)
print('*'*50)
print("Vector substraction to matrix takes advantage of broadcasting \n",mat1-vec3)
print('*'*50)
print("Dot product b/w vec and matrix via * operator \n",mat1.T*vec3)  ### why did we transpose?
print('*'*50)
print("Dot product b/w vec and matrix via @ operator  \n",mat1.T@vec3) ### why did we transpose?
print('*'*50)
print("VDot product b/w vec and matrix via np.matmul operator  \n",np.matmul(mat1.T,vec3)) ### why did we transpose?
print('*'*50)
print("Element wise multiplication takes advantage of broadcasting \n",np.multiply(mat1,vec3)) ### why did we NOT transpose?
print('*'*50)
```

***Categorical data representation via one-hot encoding & label encoder***


```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
import pandas as pd

label_enc = LabelEncoder()
Cat = ['Dog','Cat','Horse','Cat','Mouse','Goose']
Cat_encoded = label_enc.fit_transform(Cat)
dictionary = dict(zip(Cat_encoded,Cat))
print('*'*50)

# for a, b in zip(Cat, Cat_encoded):
#     print (a, b)

print([{c:a} for c, a in zip(Cat, Cat_encoded)])
print('*'*50)

OH_enc = OneHotEncoder()
Cat_OHE = pd.DataFrame(OH_enc.fit_transform(Cat_encoded.reshape(-1,1)).toarray())  ### to array method
Cat_OHE.columns = [dictionary.get(i) for i in Cat_OHE.columns]  #### .get method for dict
print('*'*50)
print(Cat_OHE)
```
